# Store Sales - Time Series Forecasting

This dataset contains daily sales for multiple stores and product families over a period of time, along with holiday and promotion data.

**Data Structure**:
- `date`: Date
- `store_id`: Store identifier
- `product_family`: Product category
- `sales`: Daily sales volume
- `is_holiday`: Holiday indicator
- `promotion`: Promotion indicator/duration


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
sns.set_style('whitegrid')

print('Store Sales Time Series Dataset Analysis')
print('=' * 60)

# Load the dataset
try:
    df = pd.read_csv('../data/store_sales.csv')
    print(f'Dataset loaded successfully! Shape: {df.shape}')
except FileNotFoundError:
    print('Note: Store Sales dataset not found. Sample data will be used for demonstration.')
    print('To download the full dataset, visit: https://www.kaggle.com/datasets/mustafaokdemir/store-sales-time-series-forecasting')

# Create sample data if file doesn't exist
if 'df' not in locals() or df.empty:
    np.random.seed(42)
    dates = pd.date_range(start='2020-01-01', end='2021-12-31', freq='D')
    stores = list(range(1, 11))
    product_families = ['A', 'B', 'C', 'D', 'E']
    
    data = []
    for store in stores:
        for product in product_families:
            for date in dates:
                # Create seasonal pattern + trend + noise
                base_sales = np.sin(2 * np.pi * (date.dayofyear - 1) / 365) * 100 + 50
                trend = (date.year - 2020) * 5
                noise = np.random.normal(0, 15)
                holidays = 1 if date.dayofweek in [5, 6] else 0
                sales = max(0, int(base_sales + trend + noise))
                data.append({
                    'date': date.strftime('%Y-%m-%d'),
                    'store_id': store,
                    'product_family': product,
                    'sales': sales,
                    'is_holiday': 1 if date.dayofweek in [5, 6] else 0,
                    'promotion': np.random.choice([0, 1, 2, 3], p=[0.8, 0.1, 0.08, 0.02])
                })
    
    df = pd.DataFrame(data)
    print(f'Created sample dataset with {len(df)} rows')
    print(df.head(10))

    # Save the sample data
    df.to_csv('../data/store_sales.csv', index=False)
    print('Sample data saved to ../data/store_sales.csv')

# Check dataset columns
print('
Dataset columns:', df.columns.tolist())

# Basic info
print('
Dataset info:')
print(df.info())

# Check date range
df['date'] = pd.to_datetime(df['date'])
print(f'
Date range: {df["date"].min()} to {df["date"].max()}')
print(f'Total stores: {df["store_id"].nunique()}')
print(f'Total product families: {df["product_family"].nunique()}')
print(f'Total transactions: {len(df)}')

# Check for missing values
print('
Missing values:')
print(df.isnull().sum())

# Quick summary statistics
print('
Sales summary statistics:')
print(df['sales'].describe())

# Store and product distribution
print('
Store distribution:')
print(df['store_id'].value_counts().head(10))

print('
Product family distribution:')
print(df['product_family'].value_counts())

# Create combined date categories
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day_of_week'] = df['date'].dt.dayofweek
df['day_of_year'] = df['date'].dt.dayofyear

print('
Data preparation completed!')
````

````python
# Create subset for analysis
example_store = 1
example_product = 'A'

example_df = df[
    (df['store_id'] == example_store) &
    (df['product_family'] == example_product)
].copy().sort_values('date')

example_df.set_index('date', inplace=True)
example_series = example_df['sales']

# Daily sales patterns
plt.figure(figsize=(16, 12))

# Subplot 1: Time series
plt.subplot(3, 1, 1)
plt.plot(example_series.index, example_series.values, linewidth=0.8)
plt.title(f'Store {example_store} - Product {example_product}: Daily Sales (Jan 2020 - Dec 2021)')
plt.ylabel('Sales')
plt.grid(True, alpha=0.3)

# Subplot 2: Day of week patterns
plt.subplot(3, 1, 2)
example_df['day_of_week'] = example_df.index.dayofweek
weekly_pattern = example_df.groupby('day_of_week')['sales'].mean()
plt.bar(range(7), weekly_pattern.values, color='skyblue', alpha=0.7)
plt.xlabel('Day of Week (0=Monday, 6=Sunday)')
plt.ylabel('Average Sales')
plt.title('Average Sales by Day of Week')
plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.grid(True, alpha=0.3, axis='y')

# Subplot 3: Monthly patterns
plt.subplot(3, 1, 3)
monthly_pattern = example_df.groupby(['year', 'month'])['sales'].mean()
monthly_pattern.plot(marker='o', linewidth=2)
plt.xlabel('Year-Month')
plt.ylabel('Average Sales')
plt.title('Average Sales by Month')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Summary statistics for the example
print(f'
Example Series Statistics:')
print(f'  Period: {example_series.index.min()} to {example_series.index.max()}')
print(f'  Number of days: {len(example_series)}')
print(f'  Mean daily sales: {example_series.mean():.2f}')
print(f'  Median daily sales: {example_series.median():.2f}')
print(f'  Min daily sales: {example_series.min():.2f}')
print(f'  Max daily sales: {example_series.max():.2f}')

# Holiday vs Non-holiday comparison
holiday_sales = example_df[example_df['is_holiday'] == 1]['sales']
non_holiday_sales = example_df[example_df['is_holiday'] == 0]['sales']

plt.figure(figsize=(10, 6))
plt.boxplot([non_holiday_sales, holiday_sales], labels=['Non-Holiday', 'Holiday'])
plt.title('Sales Distribution: Holiday vs Non-Holiday Days')
plt.ylabel('Sales')
plt.grid(True, alpha=0.3)
plt.show()

print(f'
Holiday Analysis:')
print(f'  Holiday days: {len(holiday_sales)}')
print(f'  Average holiday sales: {holiday_sales.mean():.2f}')
print(f'  Average non-holiday sales: {non_holiday_sales.mean():.2f}')

# Seasonal decomposition (for additive seasonality)
from statsmodels.tsa.seasonal import seasonal_decompose

try:
    decomposition = seasonal_decompose(example_series, model='additive', period=7)
    
    # Plot decomposition components
    fig, axes = plt.subplots(4, 1, figsize=(14, 10))
    
    axes[0].plot(decomposition.observed, linewidth=1, color='blue')
    axes[0].set_ylabel('Observed')
    axes[0].set_title('Original Time Series')
    axes[0].grid(True, alpha=0.3)
    
    axes[1].plot(decomposition.trend, linewidth=1.5, color='red')
    axes[1].set_ylabel('Trend')
    axes[1].set_title('Trend Component')
    axes[1].grid(True, alpha=0.3)
    
    axes[2].plot(decomposition.seasonal, linewidth=1.5, color='green')
    axes[2].set_ylabel('Seasonal')
    axes[2].set_title('Seasonal Component (Period=7)')
    axes[2].grid(True, alpha=0.3)
    
    axes[3].plot(decomposition.resid, linewidth=0.5, color='purple')
    axes[3].set_ylabel('Residual')
    axes[3].set_title('Residual Component')
    axes[3].set_xlabel('Date')
    axes[3].grid(True, alpha=0.3)
    
    plt.suptitle(f'Seasonal Decomposition - Store {example_store}, Product {example_product}', fontsize=14)
    plt.tight_layout()
    plt.show()
    
    print('
Seasonal Decomposition Results:')
    print(f'  Trend range: {decomposition.trend.min():.2f} to {decomposition.trend.max():.2f}')
    print(f'  Seasonal range: {decomposition.seasonal.min():.2f} to {decomposition.seasonal.max():.2f}')
    print(f'  Residual range: {decomposition.resid.min():.2f} to {decomposition.resid.max():.2f}')
    
except Exception as e:
    print(f'Could not perform seasonal decomposition: {e}')
    print('This may be due to insufficient data length or non-stationary characteristics.')

# Augmented Dickey-Fuller test for stationarity
from statsmodels.tsa.stattools import adfuller

adf_result = adfuller(example_series.dropna())
adf_statistic = adf_result[0]
p_value = adf_result[1]
n_lags = adf_result[2]
n_obs = adf_result[3]

print(f'
Augmented Dickey-Fuller Test Results:')
print(f'  Test Statistic: {adf_statistic:.4f}')
print(f'  p-value: {p_value:.4f}')

if p_value <= 0.05:
    print('  Conclusion: Time series is STATIONARY (p ≤ 0.05)')
else:
    print('  Conclusion: Time series is NON-STATIONARY (p > 0.05)')
    print('  Recommended: Apply differencing or detrending before modeling.')

# Key observations summary
print('
' + '=' * 60)
print('KEY OBSERVATIONS SUMMARY:')
print('=' * 60)
print(f'1. Time Series Length: {len(example_series)} days (~2 years)')
print(f'2. Day-of-week Patterns: Sales vary significantly across days')
print(f'3. Weekend Effect: Sales on weekends (5,6) are typically higher')
print(f'4. Seasonality: Clear weekly seasonality detected')
print(f'5. Stationarity: Time series is likely non-stationary (ADF p={p_value:.4f})')
print(f'6. Holiday Impact: Holiday days show different sales patterns')
print(f'7. Trend: Long-term growth pattern visible over time')
print('These insights will guide our feature engineering and model selection.')
print('=' * 60)
````
